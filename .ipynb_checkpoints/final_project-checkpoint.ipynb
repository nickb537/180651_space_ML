```julia
#load the data
using CSV
using DataFrames
using LinearAlgebra
using Statistics

fault_data = CSV.read("data/fault_quickscat.csv", DataFrame)
fault_data = Matrix(fault_data)
fault_data = fault_data[1:end-5,2:end]' #Remove the timestamps and transpose

normal_data = CSV.read("data/normal_quickscat.csv", DataFrame)
normal_data = Matrix(normal_data)
normal_data = normal_data[1:end-5,2:end]' #Remove the timestamps and transpose

#Each row is a parameter
#Each column is a timestamp
```

```julia
#Put it all in one big matrix
all_data = [normal_data fault_data]
normal_indices = 1:size(normal_data)[2]
fault_indices = size(normal_data)[2]+1 : size(normal_data)[2]+size(fault_data)[2]
```

```julia
plot(all_data)
```

```julia
#Want to take the average for a given param (row) and subtract it from all timestamps
#Each sample is col

using PyPlot

X = convert(Array{Float64}, all_data)

means = mean.(eachrow(X)) # subtract means from X

A = X .- means # subtract means from X

U, σ, V = svd(A);

plot(σ, "b.-")
title("Singular values of quickscat data");
xlabel(L"k");
legend([L"\sigma_k"]);
```

```julia
c1 = U[:,1]'*A
c2 = U[:,2]'*A

c=U'*A
```

```julia
figure(figsize=(20,15))
plot(c1[normal_indices][1:100:end], c2[normal_indices][1:100:end], "b.", markersize=10)
plot(c1[fault_indices][1:20:end], c2[fault_indices][1:20:end], "r.", markersize=10)

#text.(c1', c2', string.(1:length(c1)), fontsize=12);
title("Quickscat data compressed to two singular values", fontsize=20);
xlabel(L"$c_1$", fontsize=20);
ylabel(L"$c_2$", fontsize=20);
```

```julia
figure(figsize=(20,15))
sv1 = 3
sv2 = 4
plot(c[sv1,:][normal_indices][1:100:end], c[sv2,:][normal_indices][1:100:end], "b.", markersize=10)
plot(c[sv1,:][fault_indices][1:20:end], c[sv2,:][fault_indices][1:20:end], "r.", markersize=10)

#text.(c1', c2', string.(1:length(c1)), fontsize=12);
title("Quickscat data compressed to two singular values", fontsize=20);
xlabel(L"$c_3$", fontsize=20);
ylabel(L"$c_2$", fontsize=20);
```
